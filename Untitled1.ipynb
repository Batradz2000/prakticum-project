## Предобработка данных

```python
#удалим первый столбец а также приведем строки к нижнему регистру
data = data.drop(columns = data.columns[0])
data.columns = data.columns.str.lower()
```

```python
for k in ['blue_activity_count','green_gvm_order_count', 'black_gvm_order_count']:
    lists = []
    for i in data[k]:
        try:
            float(i)
        except:
            lists.append(i)
    data[k] = data[~data[k].isin(lists)][k].astype('float')
```

```python
#удалим пропуски
data.dropna()
```

```python
# из таблицы мы видим, что есть пользователи, которые не принадлежат ни одной из компании, посчитаем их количество 
data[data[['active_red','has_green_id','has_black_id','has_blue_id']].sum(axis = 1)== 0]       
```
